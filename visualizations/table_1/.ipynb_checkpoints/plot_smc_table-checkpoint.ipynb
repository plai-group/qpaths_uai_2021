```python
import wandb
import pandas as pd
api = wandb.Api()
runs = api.runs("vadmas/qpath_smc_numeric", filters={"$and": [{"state": "finished"}] })
```

```python
ground_truths = [r for r in runs if r.config['alg_type'] == 'ground_truth']
```

```python
runs = [r for r in runs if 'qpath_table' in r.tags and r.config['dataset_name'] == 'pima']
```

```python
def process_run(run):
    for f in run.files():
        if 'pickle' in f.name:
            f.download(replace=True)
            df = pd.read_pickle(f.name)
            break
    df['dataset_name'] = run.config['dataset_name']
    df['alg_type'] = run.config['alg_type']
    df['adaptive'] = run.config['adaptive']
    df['N'] = run.config['N']
    df['nruns'] = run.config['nruns']
    return df.remove_columns('out')
```

```python
res = (df[['q', 'M','type','adaptive','last_loglik','exponents']]
 .query('M < 20')
 .assign(K=lambda x:x["exponents"].apply(len))
 .drop('exponents',axis=1)
 .assign(
     err = lambda x: (x['last_loglik'] - pima_gt).abs()
 )
 .groupby(['adaptive', 'M','q']).agg(['mean','std'])
 .collapse_levels('_')
 .reset_index()
 .dropna())

geometric = res.query("q == 1")
qpath = res.query("q != 1")
```

```python
def process_best(res):
    bestq = (res
     .query("q != 1")
)
    bestq = bestq.loc[[bestq.err_mean.idxmin()]]
    return bestq
```

```python
qpath_results = qpath.groupby(['adaptive','M']).apply(process_best)
geometric_results = geometric
```
