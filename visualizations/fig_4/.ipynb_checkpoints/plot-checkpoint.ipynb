```python
pima_data = pd.read_pickle('pima_unadaptive.pickle')
```

```python
pima_data.groupby(['M','type','q']).agg(['mean','std'])
```

```python
pal = sns.cubehelix_palette(7, rot=.15, light=.8, dark=0.3)
```

```python
scale = 1.1
fig, axes = plt.subplots(1, 1, dpi=200, figsize=(30/scale, 8/scale))
sns.violinplot(x="q",
                y="last_loglik",
                hue="M", 
                data=pima_data.sort_values("q").query("q > 0.98"),
                palette=pal,
                ax=axes)
axes.axhline(
    y = -398.94903629580483,
    linewidth=1.2,
    linestyle=':',
    color='grey',
    label=f'ground truth'
)

# Put the legend out of the figure
axes.legend(loc='center left',
            bbox_to_anchor=(1, 0.5), 
            title="number of move steps",
            title_fontsize=24,
            fontsize=20,
           )

labels = [item.get_text() for item in axes.get_xticklabels()]
new_labels = [str(round(float(l), 4)) for l in labels]
new_labels[-1] = 'geometric'
axes.set_xticklabels(new_labels, fontsize=24)

axes.set_xlabel(r'$q$ (log spaced)', fontsize=24)
axes.xaxis.set_label_coords(0.5, -0.1,)


axes.set_ylabel(r'marginal likelihood', fontsize=24)

# ylabels = [-900.0, "", -700.0, "", -500.0, "", -300.0]
axes.set_yticks([-900.0, -600.0, -300.0])
axes.tick_params(axis='y', which='major', labelsize=24)
axes.yaxis.set_label_coords(-0.05,.5)
sns.despine(ax=axes, trim=False)

plot_dir = Path('.')
filename = plot_dir / f"pima_violin_plot.pdf"
fig.savefig(filename, bbox_inches='tight')
print('saved to {}'.format(filename))
```
