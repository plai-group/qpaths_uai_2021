```python
res_stud = (pd.read_pickle('./res_studt.pkl')
 .assign(
    name = lambda x: x.path + "_" + x.q.round(5).astype(str))
 .replace("qmoment_1.0",'q_moments'))
```

```python
fig, ax = plt.subplots(1, 1, dpi=200, figsize=(6, 4))

g = sns.lineplot(x='no_betas', 
             y='gap', 
             hue='name', 
             style='name', 
             data=res_stud,
             alpha=0.9,
             markers=True,
             ax=ax,
             legend=True)

ax.set_ylabel("BDMC gap (nats)")
ax.set_xlabel("K")
sns.despine(ax=ax, trim=True)
```
